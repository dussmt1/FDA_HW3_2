## imports

```python
import pandas
from matplotlib import pyplot as plt
import types
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

## helper functions

```python
def boxCompare(df, valueFunc, removeZeros=False, skipPos=0):
    if type(valueFunc) != types.LambdaType:
        raise TypeError('valueFunc should be callable')
    if removeZeros:
        tmp = valueFunc(df.loc[df.loc[:,'Revenue'] == True,:])
        tmp = tmp.loc[tmp != 0]
        plt.boxplot(tmp, positions=[skipPos+1], labels=['Revenue'])
        tmp = valueFunc(df.loc[df.loc[:,'Revenue'] == False,:])
        tmp = tmp.loc[tmp != 0]
        plt.boxplot(tmp, positions=[skipPos+2], labels=['No Revenue'])
    else:
        plt.boxplot(valueFunc(df.loc[df.loc[:,'Revenue'] == True,:]), positions=[skipPos+1], labels=['Revenue'])
        plt.boxplot(valueFunc(df.loc[df.loc[:,'Revenue'] == False,:]), positions=[skipPos+2], labels=['No Revenue'])
        
def df_avgof(a,b):
    return lambda df:(
        pandas.Series(list(map(
            lambda i: i[0]/i[1] if i[1]!=0 else i[0],
            zip(df.loc[:,a].values,df.loc[:,b].values)
        )))
    )
```

## Boxplot: revenue  

#### 覺得好像沒有什麼特別值得注意的

```python
plt.figure()
plt.title('ProductRelated')
plt.boxplot(raw.loc[raw.loc[:,'Revenue'] == True, 'ProductRelated'], positions=[1], labels=['Revenue'])
plt.boxplot(raw.loc[raw.loc[:,'Revenue'] == False, 'ProductRelated'], positions=[2], labels=['no Revenue'])
boxCompare(raw, lambda x: x.loc[:,'ProductRelated'], removeZeros=True, skipPos = 2)

plt.figure()
plt.title('ProductRelated_Duration')
boxCompare(raw, lambda x: x.loc[:,'ProductRelated_Duration'])
boxCompare(raw, lambda x: x.loc[:,'ProductRelated_Duration'], removeZeros=True, skipPos = 2)

plt.figure()
plt.figure('avg ProductRelated_Duration')
boxCompare(raw, df_avgof('ProductRelated_Duration','ProductRelated'))
boxCompare(raw, df_avgof('ProductRelated_Duration','ProductRelated'), removeZeros=True, skipPos=2)

plt.figure()
plt.title('Administrative')
boxCompare(raw, lambda x: x.loc[:,'Administrative'])
boxCompare(raw, lambda x: x.loc[:,'Administrative'], removeZeros=True, skipPos = 2)

plt.figure()
plt.title('Administrative_Duration')
boxCompare(raw, lambda x: x.loc[:,'Administrative_Duration'])
boxCompare(raw, lambda x: x.loc[:,'Administrative_Duration'], removeZeros=True, skipPos = 2)

plt.figure()
plt.title('Informational')
boxCompare(raw, lambda x: x.loc[:,'Informational'])
boxCompare(raw, lambda x: x.loc[:,'Informational'], removeZeros=True, skipPos = 2)

plt.figure()
plt.title('Informational_Duration')
boxCompare(raw, lambda x: x.loc[:,'Informational_Duration'])
boxCompare(raw, lambda x: x.loc[:,'Informational_Duration'], removeZeros=True, skipPos = 2)

plt.figure()
plt.title('BounceRates')
boxCompare(raw, lambda x: x.loc[:,'BounceRates'])
boxCompare(raw, lambda x: x.loc[:,'BounceRates'], removeZeros=True, skipPos = 2)

plt.figure()
plt.title('ExitRates')
boxCompare(raw, lambda x: x.loc[:,'ExitRates'])
boxCompare(raw, lambda x: x.loc[:,'ExitRates'], removeZeros=True, skipPos = 2)

plt.figure()
plt.title('Total Duration')
plt.ylim(top=10000)
boxCompare(raw, lambda x: x.loc[:,'Administrative_Duration']+x.loc[:,'ProductRelated_Duration']+x.loc[:,'Informational_Duration'])
boxCompare(raw, lambda x: x.loc[:,'Administrative_Duration']+x.loc[:,'ProductRelated_Duration']+x.loc[:,'Informational_Duration'], removeZeros=True, skipPos = 2)
```

```python
oldVisitorRows = raw.loc[raw.loc[:,'VisitorType']=='Returning_Visitor',:]
newVisitorRows = raw.loc[raw.loc[:,'VisitorType']=='New_Visitor',:]

print('%d old Visitor Rows\n%d new visitor rows' % (len(oldVisitorRows), len(newVisitorRows)))

plt.figure()
plt.boxplot([oldVisitorRows.loc[:,'BounceRates'], newVisitorRows.loc[:,'BounceRates']], labels=['Returning Visitor','New Visitor'])
plt.title('Bounce Rate')

print('0.2 bounce rate rows: old Visitor%d, new Visitor%d' % (sum(oldVisitorRows.loc[:,'BounceRates']==0.2),sum(newVisitorRows.loc[:,'BounceRates']==0.2)))
print('0 bounce rate rows: old Visitor%d, new Visitor%d' % (sum(oldVisitorRows.loc[:,'BounceRates']==0),sum(newVisitorRows.loc[:,'BounceRates']==0)))

plt.figure()
plt.boxplot([oldVisitorRows.loc[:,'ExitRates'], newVisitorRows.loc[:,'ExitRates']], labels=['ExitRates Visitor','New Visitor'])
plt.title('Exit Rate')

plt.figure()
plt.ylim(top=5000)
plt.boxplot([oldVisitorRows.loc[:,'ProductRelated_Duration'], newVisitorRows.loc[:,'ProductRelated_Duration']], labels=['Returning Visitor','New Visitor'])
plt.title('Product Pages time')

plt.figure()
plt.ylim(top=150)
plt.boxplot([oldVisitorRows.loc[:,'ProductRelated'], newVisitorRows.loc[:,'ProductRelated']], labels=['Returning Visitor','New Visitor'])
plt.title('Product Pages accessed')

plt.figure()
plt.boxplot([oldVisitorRows.loc[:,'Administrative'], newVisitorRows.loc[:,'Administrative']], labels=['Returning Visitor','New Visitor'])
plt.title('Administrative Pages accessed')

# print('x: ProductRelated, y: BounceRates, red:oldVisitor, blue: newVisitor')
plt.figure()
plt.scatter(oldVisitorRows.loc[:,'ProductRelated'],oldVisitorRows.loc[:,'BounceRates'],c='red')
plt.figure()
plt.scatter(newVisitorRows.loc[:,'ProductRelated'],newVisitorRows.loc[:,'BounceRates'],c='blue')

# print('x: ExitRates, y: BounceRates, red:oldVisitor, blue: newVisitor')
plt.figure()
plt.scatter(oldVisitorRows.loc[:,'ExitRates'],oldVisitorRows.loc[:,'BounceRates'],c='red')
plt.figure()
plt.scatter(newVisitorRows.loc[:,'ExitRates'],newVisitorRows.loc[:,'BounceRates'],c='blue')
```

## find zeros  

#### 因為上面發現Bounce Rate有很多0  
#### 所以來看一下各欄位0的數量  
> Month, OS, Browser, Region, TrafficType, VisitorType, Weekend, Revenue  
> 總是有值的class  

> 存在大量0的數值欄位:  
> Informational(\_Duration) 
>>  可能大多數根本沒看 informational page
>
> PageValues
>> 不知道這是甚麼
> 
> SpecialDay
>> 不是每天都過節

```python
cols = raw.columns

print('rows: %d' % (len(raw)))

for x in cols:
    if type(raw.loc[0,x]) in [np.int64, np.float64]:
        print('zeros found in column %s: %d' % (x, sum(raw.loc[:,x] == 0)))
    elif type(raw.loc[0,x]) in [np.bool_]:
        print('False found in column %s: %d' % (x, sum(raw.loc[:,x] == False)))
    elif type(raw.loc[0,x]) in [str]:
        print('Empty found in column %s: %d' % (x, sum(raw.loc[:,x] == '')))
    else:
        print('column with unexpected datatype: %s' % (x))
```

## 預測

#### 可以用其他的column來猜客戶是不是new visitor  
先用感覺有差的BounceRates跟ExitRates, 還有0出現率低的ProductRelated  
來試試看

```python
raw.loc[:, 'is_new'] = raw.loc[:,'VisitorType'] == 'New_Visitor'

model1 = KNeighborsClassifier()
use_cols1 = ['ProductRelated', 'BounceRates', 'ExitRates']

testRows = list(range(len(raw)))
testRows = list(map(lambda x: 1 if x%10==3 else 0, testRows))
trainRows = list(map(lambda x: 1-x, testRows))

model1.fit(raw.loc[trainRows, use_cols1], raw.loc[trainRows, 'is_new'])

g = model1.predict(raw.loc[testRows, use_cols1])
print(g)

print('acc: %f' % (sum(g == raw.loc[testRows, 'is_new'])/sum(testRows)))
```
